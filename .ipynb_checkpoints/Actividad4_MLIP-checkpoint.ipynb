### 1 Binarización: Clasificación dentro de una imagen en 2 clases, utilizando una variable: intensidad

Cargar en la notebook utilizando la librería OpenCV una imagen de manera monocroma que contenga un objeto claro u oscuro a destacar del fondo y luego:

* Binarizar la imagen por umbralado de la intensidad (utilizar el histograma de ser necesario para la elección del umbral), buscando separar el objeto deseado del fondo (unos,255 o rue en el objeto y ceros o False en el fondo).
* Mejorar la binarización con un filtro morfológico que crea adecuado.
* Encontrar los bordes del objeto utilizando operaciones morfológicas o gradiente morfológico.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('Union estructural c.jpg',cv2.IMREAD_GRAYSCALE)
print('tamaño de la imagen monocroma:', img.shape)
plt.imshow(img, cmap='gray')
plt.colorbar()
plt.show()
```

```python
print('Intensidades de toda la imagen monocroma:')
print ('Media:', np.mean(img),', Mediana:', np.median(img),
       ', Desvío:', np.std(img))

plt.hist(img.ravel(), 256,[0,256] ) 
plt.title('histograma'), plt.xlabel('bins de intensidad de iluminacion'), plt.ylabel('cantidad de pixeles')
plt.show()
```

## Binarizacion

```python
#Quitamos ruido y aplicamos Sobel
blur = cv2.GaussianBlur(img,(9,9),0)
sobelx = cv2.Sobel(src = blur, ddepth = cv2.CV_64F, dx = 1, dy = 0, ksize = 3) 
sobely = cv2.Sobel(blur, cv2.CV_64F, dx = 0, dy = 1, ksize = 3)                                  
res=np.sqrt(sobelx ** 2 + sobely ** 2)
arr = np.uint8(res)


#Binarizamos
ret3,binaria3 = cv2.threshold(arr,0,255,cv2.THRESH_OTSU)#Binarizacion automatica por OTSU
print('Binarizacion automatica por OTSU',ret3)
ret1,binaria1 = cv2.threshold(res,60,255,cv2.THRESH_BINARY)
print('Binarizacion',ret1)


fig, axs = plt.subplots(nrows=2, ncols=2, sharex=False)
axs=axs.ravel()
fig.set_figheight(6)
fig.set_figwidth(10)


axs[0].imshow(binaria3, cmap='gray',vmin=0 ,vmax=255)
axs[0].set_title('Imagen binaria OTZU')

axs[1].imshow(binaria1, cmap='gray',vmin=0 ,vmax=255)
axs[1].set_title('Imagen binaria ')

axs[2].hist(binaria3[:,:].ravel(), 256,[0,256] ) 
axs[2].set_xlabel('bins de intensidad')
axs[2].set_ylabel('cantidad de pixeles')

axs[3].hist(binaria1[:,:].ravel(), 256,[0,256] ) 
axs[3].set_xlabel('bins de intensidad')
axs[3].set_ylabel('cantidad de pixeles')

plt.show()

```

## Morfologia

```python
#Elegimos la imagen binarizada (binaria1) con el umbral de 60
kernel=cv2.getStructuringElement(cv2.MORPH_RECT, (5, 3))

cierre=cv2.morphologyEx(binaria1,cv2.MORPH_CLOSE,kernel)
apertura = cv2.morphologyEx(binaria1, cv2.MORPH_OPEN, kernel)

fig = plt.gcf()
fig.set_size_inches(14.5, 10.5)
plt.subplot(1,3,1), plt.imshow(binaria1,cmap='gray'),plt.title('Imagen Binaria Original'),plt.xticks([]), plt.yticks([])
plt.subplot(1,3,2),plt.imshow(apertura,cmap='gray'),plt.title('Apertura'),plt.xticks([]), plt.yticks([])
plt.subplot(1,3,3),plt.imshow(cierre,cmap='gray'),plt.title('Cierre'),plt.xticks([]), plt.yticks([])

plt.show()
```

## Extraemos bordes

```python
kernel=cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))

erosion = cv2.erode(binaria1,kernel,iterations = 1)

dilatacion = cv2.dilate(binaria1,kernel,iterations = 1)

gradiente=cv2.morphologyEx(binaria1,cv2.MORPH_GRADIENT,kernel)

fig = plt.gcf()
fig.set_size_inches(14.5, 10.5)
plt.subplot(1,4,1), plt.imshow(binaria1,cmap='gray'),plt.title('Imagen Original'),plt.xticks([]), plt.yticks([])
plt.subplot(1,4,2),plt.imshow(binaria1-erosion,cmap='gray'),plt.title('Borde interior'),plt.xticks([]), plt.yticks([])
plt.subplot(1,4,3),plt.imshow(dilatacion-binaria1,cmap='gray'),plt.title('Borde exterior'),plt.xticks([]), plt.yticks([])
plt.subplot(1,4,4),plt.imshow(gradiente,cmap='gray'),plt.title('Gradiente'),plt.xticks([]), plt.yticks([])
plt.show()
```

### 2 Clasificación dentro de una imagen: 4 clases, 3 variables o features: R,G,B

Cargue una imagen RGB a elección que a la vista tenga 4 clases identificables:

* Describa con sus palabras brevemente las 4 clases que ve en la imagen.
* Utilice k-medias con k=4 cantidad de clusters para realizar clasificación no supervisada de la imagen. Muestre los resultados, tanto en el espacio RGB como en la disposición de los clusters encontrados en la imagen.
* Realice una clasificación semi-supervisada: 
 - etiquetando manualmente pixeles de las 4 clases. muestre los pixeles etiquetados en el espacio RGB. 
 - reclasificando toda la imagen según esas cuatro clases.
 - Muestre los resultados y realice una mejora de la clasificación utilizando un filtro de moda
 * Comente los resultados. 
 
 Ayuda: Utilice la distancia euclídea. Normalice las observaciones si lo requiere su conjunto de datos ("distancia del coseno") 

```python
import cv2
satelite_cv=cv2.imread('satelite.jpg')
satelite=satelite_cv.copy()
satelite[:,:,[2,1,0]]=satelite_cv #De BGR a RGB

from matplotlib import pyplot as plt
plt.imshow(satelite)
plt.show()  
```

Las clases que se observan en la imagen son:
-La ciudad, de un color gris
-El oceano, de color azul
-Las monteñas, de color marron
-El romper de las olas, color blanco

```python
import numpy as np
dims=satelite.shape
cant_filas=dims[0]
cant_columnas=dims[1]
cant_pixeles=cant_filas*cant_columnas
data0=np.zeros([cant_pixeles,4],dtype='int')
for f in range(cant_filas):
    for c in range(cant_columnas):
        nro=f*cant_columnas+c
        data0[nro,0]=nro
        data0[nro,1:4]=satelite[f,c,:]

import pandas as pd
df = pd.DataFrame(data0, columns=["pixel_id","rojo","verde","azul"])

print(df.shape)
print('Base de pixeles:')
df.head()  
```

```python
from sklearn.cluster import KMeans
from sklearn import preprocessing
from sklearn.metrics import silhouette_score

df_pix=df.drop('pixel_id',axis=1) 
kmedias3=KMeans(n_clusters=3).fit(df_pix)

print('Suma de los cuadrados de las distancias al clusters / Inertia: ', kmedias3.inertia_)
```

```python
scores = [KMeans(n_clusters=i+2).fit(df_pix).inertia_ for i in range(10)]
plt.plot(np.arange(2, 12), scores)
plt.xlabel('Number of clusters')
plt.ylabel("Inertia")
plt.title("Inertia of k-Means versus number of clusters")
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(df_pix)
df_clusters=df.copy()
df_clusters['kmedias_4']=kmeans.labels_
df_clusters.head()
```

## Resultados en espacio RGB

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

colores=['purple','blue','orange','red']
variables=["rojo","verde","azul"]

variable1=variables[1]
variable2=variables[0]

f1 = df_clusters[variable1].values
f2 = df_clusters[variable2].values
f3 = df_clusters.kmedias_4

asignar=[]
l=len(f3)
print('se grafican de a dos (dimensión 2)')

for k in range(l): 
    asignar.append(colores[f3[k]-1])
    
plt.scatter(f1, f2, c=asignar, s=1)
plt.xlabel(variable1)
plt.ylabel(variable2)
patch1 = mpatches.Patch(color=colores[0], label='cluster0')
patch2 = mpatches.Patch(color=colores[1], label='cluster1')
patch3 = mpatches.Patch(color=colores[2], label='cluster2')
patch4 = mpatches.Patch(color=colores[3], label='cluster3')
plt.legend(handles=[patch1,patch2,patch3,patch4])
plt.title('Kmedias y Distancia euclídea en espacio RGB')
plt.show()
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(df[variables[0]].values, df[variables[1]].values, df[variables[2]].values,c=asignar, s=1)
ax.set_xlabel(variables[0])
ax.set_ylabel(variables[1])
ax.set_zlabel(variables[2])
plt.show()
```

## Clasificacion semi supervisada

```python
import cv2
satelite_cv=cv2.imread('satelite.jpg')#al abrir con cv2 se abre como BGR
print(satelite_cv.size)
print(satelite_cv.shape)


satelite=satelite_cv.copy()
satelite[:,:,[2,1,0]]=satelite_cv

from matplotlib import pyplot as plt
plt.imshow(satelite)
plt.show()
```

```python
import numpy as np
dims=satelite.shape
cant_filas=dims[0]
cant_columnas=dims[1]
cant_pixeles=cant_filas*cant_columnas
data0=np.zeros([cant_pixeles,4],dtype='int')
for f in range(cant_filas):
    for c in range(cant_columnas):
        nro=f*cant_columnas+c
        data0[nro,0]=nro
        data0[nro,1:4]=satelite[f,c,:]
        
import pandas as pd
df = pd.DataFrame(data0, columns=["pixel_id","rojo","verde","azul"])

print(df.shape)
print('Base de pixeles:')
df.head(171961) 
```

```python
#ciudad
showCrosshair = False
fromCenter = False
roi1 = cv2.selectROI("delimite la ROI con mouse +ent+tecla", satelite_cv, fromCenter, showCrosshair)
cv2.waitKey() #espera pulsado de tecla. 
print('ROI1:',roi1)
print('(columna, fila, anchocol, anchofila)')
cv2.destroyAllWindows()
```

```python
#Montañas
showCrosshair = False
fromCenter = False
roi2 = cv2.selectROI("delimite la ROI con mouse+ent+tecla", satelite_cv, fromCenter, showCrosshair)
cv2.waitKey() #espera pulsado de tecla. 
print('ROI2:',roi2)
print('(columna, fila, anchocol, anchofila)')
cv2.destroyAllWindows()
```

```python
#Oceano
showCrosshair = False
fromCenter = False
roi3 = cv2.selectROI("delimite la ROI con mouse+ent+tecla", satelite_cv, fromCenter, showCrosshair)
cv2.waitKey() #espera pulsado de tecla. 
print('ROI3:',roi3)
print('(columna, fila, anchocol, anchofila)')
cv2.destroyAllWindows()
```

```python
rois=[roi1,roi2,roi3]

import numpy as np
cantidad=0
for i in range(3):
    cantidad=cantidad+(rois[i][2]*rois[i][3])
print (cantidad)
data=np.zeros([cantidad,5],dtype='int')
target=np.zeros(cantidad,dtype='int')
cuenta=0
for i in range(3):
    roi=rois[i]
    
    for fil in range (roi[1],roi[1]+roi[3]):
        for col in range (roi[0],roi[0]+roi[2]):
            nro=fil*cant_columnas+col
            data[cuenta,0]=nro
            data[cuenta,1:4]=satelite[fil,col,:]
            data[cuenta,4]=i+1
            target[cuenta]=i+1
            cuenta=cuenta+1

import pandas as pd
df_etiquetado = pd.DataFrame(data, columns=["pixel_id","rojo","verde","azul","etiqueta"])
print(df_etiquetado.shape)
print('Base de pixeles etiquetados:')
df_etiquetado.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

colores=['green','blue','pink']
variables=["rojo","verde","azul"]

variable1=variables[1]
variable2=variables[0]

f1 = df_etiquetado[variable1].values
f2 = df_etiquetado[variable2].values
f3 = target

# Vamos a pintar en colores las 3 clases: verde: 0, azul:1, rosa:2
asignar=[]
l=len(f3)
print('cantidad de muestras etiquetadas:',l)
print('cantidad de clases etiquetadas:',3)
print('cantidad de variables/dimension:',3)
print('se grafican de a dos (dimensión 2)')

for k in range(l): #for index, row in filtered_data.iterrows():
    if(f3[k]==1):
        asignar.append(colores[0])
    elif (f3[k]==2):
        asignar.append(colores[1])
    else:
        asignar.append(colores[2])
        
plt.scatter(f1, f2, c=asignar, s=1)
plt.xlabel(variable1)
plt.ylabel(variable2)
patch1 = mpatches.Patch(color=colores[0], label='ciudad')
patch2 = mpatches.Patch(color=colores[1], label='montaña')
patch3 = mpatches.Patch(color=colores[2], label='oceano')
plt.legend(handles=[patch1,patch2,patch3])
plt.show()
```

```python
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D

fig = pyplot.figure()
ax = Axes3D(fig)

ax.scatter(df_etiquetado[variables[0]].values, df_etiquetado[variables[1]].values, df_etiquetado[variables[2]].values,c=asignar, s=1)
ax.set_xlabel(variables[0])
ax.set_ylabel(variables[1])
ax.set_zlabel(variables[2])
plt.show()

#pyplot.show()
```

* Utilice **dos** redes neuronales convolucionales para clasificar cada imagen en los 10 tipos de dígitos y compárelas en cuanto a su complejidad (cantidad de parámetros) y overfitting. Comente resultados

```python
### UTILIZAR TANTAS CELDAS DE CÓDIGO COMO SEA NECESARIO
```

```python
import tensorflow
from tensorflow import keras
from keras.utils import to_categorical
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import os
import cv2
import random
import numpy as np
from PIL import Image
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()

print(X_train.shape)
print(X_test.shape)

unicos, cantidad = np.unique(y_test, return_counts = True)
print(f'El dataset posee imágenes de los siguentes números: {unicos}, y a cada número le corresponde la siguente cantidad: {cantidad}')

unicos, cantidad = np.unique(y_test, return_counts = True)
print(f'El dataset posee imágenes de los siguentes números: {unicos}, y a cada número le corresponde la siguente cantidad {cantidad}')
```

```python
#Procesamiento de set de datos

X_train = X_train.reshape(60000, 784)
X_test = X_test.reshape(10000, 784)
```

```python
 #números enteros a "One - Hot" 
y_train = keras.utils.to_categorical(y_train)
y_test = keras.utils.to_categorical(y_test)
```

```python
#Estructura Neuronal A
# capa de entrada 
Imagen_de_entrada = layers.Input(shape = (784,))

# Primera Capa oculta 
Capa1 = layers.Dense(units = 350, activation = 'relu')(Imagen_de_entrada)

# Segunda capa oculta 
Capa2 = layers.Dense(units = 225, activation = 'relu')(Capa1)

#Tercera capa oculta 
Capa3 = layers.Dense(units = 200, activation = 'relu')(Capa2)

# Capa de Salida 
Salida = layers.Dense(units = 10, activation = 'softmax')(Capa3)

#Modelo de Keras
Red_Neuronal = keras.Model(inputs = Imagen_de_entrada, outputs = Salida)

#compilacion del modelo
Red_Neuronal.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

Historia = Red_Neuronal.fit(x = X_train, y = y_train, epochs = 10, batch_size=32, validation_data = (X_test, y_test))
```

```python
plt.plot(Historia.history['accuracy'], label = 'Entrenamiento')
plt.plot(Historia.history['val_accuracy'], label = 'Prueba')
plt.grid()
plt.xlabel('Cantidad de Épocas')
plt.ylabel('Accuracy')
plt.legend()
plt.show()


Red_Neuronal.summary()
```

```python
#Estructura Neuronal B
# capa de entrada 
Imagen_de_entrada = layers.Input(shape = (784,))

# Primera Capa oculta 
Capa1 = layers.Dense(units = 64, activation = 'relu')(Imagen_de_entrada)

# Segunda capa oculta 
Capa2 = layers.Dense(units = 128, activation = 'relu')(Capa1)

# Capa de Salida 
Salida = layers.Dense(units = 10, activation = 'softmax')(Capa2)

#Modelo de Keras
Red_Neuronal = keras.Model(inputs = Imagen_de_entrada, outputs = Salida)

#compilacion del modelo
Red_Neuronal.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

Historia = Red_Neuronal.fit(x = X_train, y = y_train, epochs = 10, batch_size=32, validation_data = (X_test, y_test))
```

```python
plt.plot(Historia.history['accuracy'], label = 'Entrenamiento')
plt.plot(Historia.history['val_accuracy'], label = 'Prueba')
plt.grid()
plt.xlabel('Cantidad de Épocas')
plt.ylabel('Accuracy')
plt.legend()
plt.show()


Red_Neuronal.summary()
```

Haciendo uso de dos redes totalmente conectadas de diferente cantidad de capas y neuronas, para clasificar cada imagen en los 10 tipos de dígitos. Al compararlas se puede ver que el rendimiento del modelo A, al tener un mayor numero de capas y neuronas,puede adaptarce mas facilmente al conjunto de entrenamiento.Pero vemos que el rendimiento de la metrica del conjunto de prueba va disminuyendo fuertemente respecto al conjunto de entrenamiento. Ademas el numero de parametros supera ampliamente la cantidad de objetos usados para el entrenamiento. Por lo que podemos decir que el modelo A esta claramente sobreajustado.  Por otro lado en el modelo B el rendimiento del conjunto de entrenamiento y de prueba tienen una pequeña difrencia pero se mueven en la misma direccion. Diciendonos entoces que el modelo B es practicamente correcto.

## Redes neuronales convolucionales

```python
#Caraga de datos
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
#Procesamiento de datos
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
#Estructura Neuronal A
# capa de entrada 
Imagen_de_entrada = layers.Input(shape = (28, 28, 1))
#Capa de combolucion
X = layers.Conv2D(filters = 32, kernel_size = (3, 3), activation = None)(Imagen_de_entrada)
#Capa de Normalizacion
X = layers.BatchNormalization()(X)
#Capa de activacion
X = layers.Activation("relu")(X)

#Capa de Max Pooling
X = layers.MaxPooling2D(pool_size = (2, 2))(X)

#Definicion de otra capa de filtrado
X = layers.Conv2D(filters = 64, kernel_size = (3, 3), activation = None)(X)

X = layers.BatchNormalization()(X)
  
X = layers.Activation("relu")(X)

# Capa Flatten
Flatten = layers.Flatten()(X)

#Capa red fully conected
X = layers.Dense(256, activation = "sigmoid")(Flatten)

X = layers.Dense(64, activation = "sigmoid")(X)

Salida = layers.Dense(units = 10, activation = 'softmax')(X)

#Modelo Keras
Red_Neuronal = keras.Model(inputs = Imagen_de_entrada, outputs = Salida)

#compilacion del modelo
Red_Neuronal.compile(loss = 'categorical_crossentropy', optimizer = 'Adam', metrics = ['accuracy'])

Historia = Red_Neuronal.fit(x = X_train, y = y_train, epochs = 10, batch_size = 32, validation_data = (X_test, y_test))
```

```python
plt.plot(Historia.history['accuracy'], label = 'Entrenamiento/train')
plt.plot(Historia.history['val_accuracy'], label = 'Prueba/test')
plt.grid()
plt.xlabel('Cantidad de Épocas')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

Red_Neuronal.summary()
```

## Regularizacion

```python
#Caraga de datos
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
print(X_train.shape, y_train.shape)
```

```python
#Funcion de rotacion
def rotacion(imagen, angulo):
    imagen = Image.fromarray(imagen)
    imagen_rotada = imagen.rotate(angle = angulo, expand = False)
    return np.array(imagen_rotada)
```

```python
#Aumento de datos
def aumentacion(imagenes, etiquetas):
    lista_imagenes = []
    lista_etiquetas = []
    
    for i in range(imagenes.shape[0]):
        
        imagen, etiqueta = imagenes[i], etiquetas[i]

        lista_imagenes.append(imagen)
        lista_etiquetas.append(etiqueta)

        lista_imagenes.append(rotacion(imagen, 5))
        lista_etiquetas.append(etiqueta)    

        lista_imagenes.append(rotacion(imagen, -5))
        lista_etiquetas.append(etiqueta)
        
    list_aux = list(zip(lista_imagenes, lista_etiquetas))

    random.shuffle(list_aux)

    lista_imagenes, lista_etiquetas = zip(*list_aux)
    
    return np.asarray(lista_imagenes), np.asarray(lista_etiquetas)
```

```python
#Aumento de datos
X_train, y_train = aumentacion(X_train, y_train)
#Tamaño de la base de datos
print(X_train.shape, y_train.shape)
#Formato One HOT
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
#Estructura Neuronal 
# capa de entrada 
Imagen_de_entrada = layers.Input(shape = (28, 28, 1))
#Capa de combolucion
X = layers.Conv2D(filters = 32, kernel_size = (3, 3), activation = None)(Imagen_de_entrada)
#Capa de Normalizacion
X = layers.BatchNormalization()(X)
#Capa de activacion
X = layers.Activation("relu")(X)

#Capa de Max Pooling
X = layers.MaxPooling2D(pool_size = (2, 2))(X)

#Definicion de otra capa de filtrado
X = layers.Conv2D(filters = 64, kernel_size = (3, 3), activation = None)(X)

X = layers.BatchNormalization()(X)
  
X = layers.Activation("relu")(X)

# Capa Flatten
Flatten = layers.Flatten()(X)

#Capa red fully conected
X = layers.Dense(256, activation = "sigmoid")(Flatten)

X = layers.Dense(64, activation = "sigmoid")(X)

Salida = layers.Dense(units = 10, activation = 'softmax')(X)

#Modelo Keras
Red_Neuronal = keras.Model(inputs = Imagen_de_entrada, outputs = Salida)

#compilacion del modelo
Red_Neuronal.compile(loss = 'categorical_crossentropy', optimizer = 'Adam', metrics = ['accuracy'])

Historia = Red_Neuronal.fit(x = X_train, y = y_train, epochs = 10, batch_size = 32, validation_data = (X_test, y_test))
```
